```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("data/iai4i2020raw.csv")

df.head()
```

```python
df_clean = df.copy()
```

```python
df_clean.drop(
    columns=["UDI","Product ID"],
    inplace=True
)
```

```python
df_clean.drop(
    columns=["TWF","HDF","PWF","OSF","RNF"],
    inplace=True
)
```

```python
df_clean.columns
```

```python
df_clean["Type"].unique()
```

```python
# Type column kept as-is (string: L / M / H)
# OneHotEncoder in the sklearn pipeline handles encoding.
# LabelEncoder was removed to avoid mismatch with training pipeline.
print(df_clean['Type'].unique())
```

```python
df_clean["Type"].unique()
```

```python
df_clean.info()
```

```python
df_clean["Wear_Speed_Ratio"] = (
    df_clean["Tool wear [min]"] /
    df_clean["Rotational speed [rpm]"]
)
```

```python
df_clean["Temp_Diff"] = (
    df_clean["Process temperature [K]"]
    -
    df_clean["Air temperature [K]"]
)
```

```python
df_clean["Power_Index"] = (
    df_clean["Torque [Nm]"]
    *
    df_clean["Rotational speed [rpm]"]
)
```

```python
df_clean["Temperature_Stress"] = (
    df_clean["Process temperature [K]"]
    *
    df_clean["Torque [Nm]"]
)
```

```python
df_clean.head()
```

```python
df_clean.to_csv("data/cleanData.csv" ,index = False)
```
